```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# UserDefined 11 values distribution

```python
nList=[11,22,33,44,1,2,3,4,5,6,6,7,7,8,9,10,12,13,15,16] #continous value dataset
dataset=pd.DataFrame(nList,columns=['x'])
```

```python
import scipy.stats
dataset['pdfCol']=scipy.stats.norm(dataset.x.mean(),dataset.x.std()).pdf(dataset.x)
plt.scatter(dataset.x,dataset.pdfCol)
plt.show()
```
